```python
var("x,y");
```

```python
eq1=5*x+9*y-1
```

```python
eq2=-x-y+2
```

```python
sol=solve([eq1,eq2],x,y);sol
```

```python
px=sol[0][0].rhs();px
```

```python
py=sol[0][1].rhs();py
```

```python
plot=plot_vector_field((eq1,eq2), (x,px-7,px+7), (y,py-7,py+7));plot
```

```python
A=matrix([[diff(eq1,x),diff(eq1,y)],[diff(eq2,x),diff(eq2,y)]]).subs(x=px,y=py);A
```

```python
A.eigenvalues()
```

Ens trobem amb dos vaps reals iguals positius.Tenim un NODE, no caldria dibuixar les separatrius


```python
vec=A.eigenvectors_right();vec
```

```python
h=0.01 # farem un petit desplaçament del punt singular en la direcció del vector propi
step1=0.01 # El pas de integració
kt=2.6;
if(vec[0][0]>0):
    sol=desolve_system_rk4([eq1, eq2],[x,y],ics=[0,px+h*vec[0][1][0][0],py+h*vec[0][1][0][1]],ivar=t,
                       end_points=[px+kt,py+kt],step=step1)
elif (vec[0][0]<0):
    sol=desolve_system_rk4([-eq1,-eq2],[x,y],ics=[0,px+h*vec[0][1][0][0],py+h*vec[0][1][0][1]],ivar=t,
                       end_points=[px+kt,py+kt],step=step1)
sol1=[[j,k] for i,j,k in sol]
if(vec[0][0]<0):
    sep1=list_plot(sol1,color="blue")
elif (vec[0][0]>0):
    sep1=list_plot(sol1,color="red")
show(sep1)
```

```python
kt=20;
if(vec[0][0]>0):
    sol=desolve_system_rk4([eq1, eq2],[x,y],ics=[0,px-h*vec[0][1][0][0],py-h*vec[0][1][0][1]],ivar=t,
                       end_points=[px+kt,py+kt],step=step1)
elif (vec[0][0]<0):
    sol=desolve_system_rk4([-eq1,-eq2],[x,y],ics=[0,px-h*vec[0][1][0][0],py-h*vec[0][1][0][1]],ivar=t,
                       end_points=[px+kt,py+kt],step=step1)
sol1=[[j,k] for i,j,k in sol]
if(vec[0][0]<0):
    sep2=list_plot(sol1,color="blue")
elif (vec[0][0]>0):
    sep2=list_plot(sol1,color="red")
show(sep2)
```

```python
var("t");
x1=function("x1")(t);
y1=function("y1")(t);
eqd1=diff(x1(t),t)==eq1.subs(x=x1(t),y=y1(t));
eqd2=diff(y1(t),t)==eq2.subs(x=x1(t),y=y1(t));
eqd2
```

```python
sol=desolve_system([eqd1, eqd2],[x1,y1],ics=[0,px+0.5,py+0.5]);sol
```

```python
plot1=parametric_plot( (sol[0].rhs(),sol[1].rhs() ),(t,-1.2,.38),color="yellow")
```
